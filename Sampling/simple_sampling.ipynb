```python
from scipy.stats import multivariate_normal
from pprint import pprint
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def uniform(min_range, max_range, sample_size):
    sample = []
    for i in range(sample_size):
        sample.append(min_range + np.random.rand() * (max_range - min_range))
    return sample
```

```python
def gaussian(mu, sigma, sample_size):
    sample = []
    uniform_distribution = uniform(1, 6, sample_size)
    y = multivariate_normal.pdf(uniform_distribution, mean=mu, cov=sigma)
    for i in range(len(y)):
        if max(y) * np.random.rand() < y[i]:
            sample.append(uniform_distribution[i])
    return sample
```

```python
def gaussian_2d(mu, sigma, sample_size):
    sample = []
    i = 0
    while i < sample_size:
        new_point = [np.random.uniform(0,10), np.random.uniform(0,10)]
        p = multivariate_normal.pdf(new_point, mean=mu, cov=sigma)
        if np.random.uniform(0,1) <= p:
            sample.append(new_point)
            i += 1
    return sample
```

```python
uniform_distribution = uniform(1, 6, 1000)
plt.hist(uniform_distribution)
plt.show()
```

```python
gaussian_distribution = gaussian(3, 1, 1000)
plt.hist(gaussian_distribution)
plt.show()
```

```python
gaussian_distribution = gaussian_2d([5,5], [[0.8,0.4], [0.4,0.8]], 500)
for x,y in gaussian_distribution:
    plt.scatter(x,y)
plt.show()
```
